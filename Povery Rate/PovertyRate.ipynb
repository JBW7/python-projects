## Section 2: Exploratory Data Analysis

I want to get a general overview of all 3 parameters of this dataset. By analyzing its mean, median, standard deviation, maximum, minimum, and range of each parameter. This will help me identify if the data is within a general range and if any states are outliers. This insight will be super helpful for the next part of my project.

```python
# Poverty Rate
pr_mean = df.PovertyRate.mean()
print ("Poverty Rate mean = " + str(pr_mean))

pr_median = df.PovertyRate.median()
print ("Poverty Rate median = " + str(pr_median))

pr_std = df.PovertyRate.std()
print ("Poverty Rate standard deviation = " + str(pr_std))

pr_max = df.PovertyRate.max()
print ("Poverty Rate max = " + str(pr_max))

pr_min = df.PovertyRate.min()
print ("Poverty Rate min = " + str(pr_min))

pr_range = pr_max - pr_min
print ("Poverty Rate range = " + str(pr_range))

```

##### Poverty Rate analysis
The mean and median poverty rates are quite close at around 12.57 and 12.1 respectively, indicating a fairly symmetrical distribution of poverty rates across the states. However, the standard deviation is about 2.65, which suggests a moderate spread of values, and the range of 12.1 percent to significant disparities in poverty rates between different states.

```python
# Unemployment Rate
ur_mean = df.UnemploymentRate.mean()
print ("Unemployment Rate mean = " + str(ur_mean))

ur_median = df.UnemploymentRate.median()
print ("Unemployment Rate median = " + str(ur_median))

ur_std = df.UnemploymentRate.std()
print ("Unemployment Rate standard deviation = " + str(ur_std))

ur_max = df.UnemploymentRate.max()
print ("Unemployment Rate max = " + str(ur_max))

ur_min = df.UnemploymentRate.min()
print ("Unemployment Rate min = " + str(ur_min))

ur_range = ur_max - ur_min
print ("Unemployment Rate range = " + str(ur_range))
```

##### Unemployment rate analysis 
The unemployment rate's mean and median are close to each other, at about 4.83 and 4.9, suggesting a relatively even distribution of unemployment rates across the states. The range of 4.7%, along with a standard deviation of about 1.25, slight disparities in unemployment rates, therefore we can make a conclusion that there is a certain level of consistency in the job markets across different states.

```python
# Median Household Income
mh_mean = df.MedianHouseholdIncome.mean()
print ("Median Household Income mean = " + str(mh_mean))

mh_median = df.MedianHouseholdIncome.median()
print ("Median Household Income median = " + str(mh_median))

mh_std = df.MedianHouseholdIncome.std()
print ("Median Household Income standard deviation = " + str(mh_std))

mh_max = df.MedianHouseholdIncome.max()
print ("Median Household Income max = " + str(mh_max))

mh_min = df.MedianHouseholdIncome.min()
print ("Median Household Income min = " + str(mh_min))

mh_range = mh_max - mh_min
print ("Median Household Income range = " + str(mh_range))
```

##### Median Household Income analysis

The mean Median Household Income is higher than the median, hinting at a distribution that is slightly skewed by higher incomes in some states. With a large range of $42,201 and a standard deviation of $11,320, there's a notable variation in income levels across states, indicating economic diversity.

##### Summary
The analysis of poverty, unemployment rates, and median household income across the states reveals a economy where economic indicators are relatively stable, yet there are still disparities. The tight clustering of unemployment rates suggests consistent job markets, while the wider spread in household incomes points to varying degrees of economic well-being. The proximity of mean and median values in all three categories indicates no extreme skewness but highlights the presence of economic inequality among states.
